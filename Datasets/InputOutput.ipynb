```python
import csv
grades = []
f = open('gradebook.csv', 'r')
for row in csv.reader(f, delimiter=','):
    grades.append(row)
    
grades = grades[1:]  # remove 1st row

for row in grades:
    row[4] = float(row[4])
    
print(grades)
f.close()
```
